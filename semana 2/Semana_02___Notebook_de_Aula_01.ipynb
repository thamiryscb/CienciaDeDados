## Etapa 3 — Estrutura mínima do projeto

Um projeto de dados minimamente organizado costuma separar:

- **raw**: dados brutos, sem alteração;
- **interim**: dados intermediários;
- **processed**: dados mais preparados para análise;
- **docs**: documentação;
- **notebooks**: experimentos e análises.

Na aula de hoje, vamos apenas montar a estrutura.

```python
project_dir = BASE_DIR / 'projeto_dados_semana2'

for sub in ['data/raw', 'data/interim', 'data/processed', 'docs', 'notebooks']:
    (project_dir / sub).mkdir(parents=True, exist_ok=True)

print('Estrutura criada em:', project_dir.resolve())
for p in sorted(project_dir.rglob('*')):
    if p.is_dir():
        print(p.relative_to(project_dir))
```

## Etapa 7 — Dicionário preliminar de variáveis

Escolham uma base e comecem a construir um dicionário preliminar. Depois ampliem para as outras.

```python
def criar_dicionario(df, nome_base):
    return pd.DataFrame({
        'base': nome_base,
        'variavel': df.columns,
        'descricao_inferida': ['' for _ in df.columns],
        'tipo_aparente': [str(df[c].dtype) for c in df.columns],
        'exemplo': [df[c].dropna().iloc[0] if df[c].dropna().shape[0] > 0 else None for c in df.columns],
        'observacoes': ['' for _ in df.columns],
    })

dicionario_labs = criar_dicionario(labs, 'laboratorios.xlsx')
dicionario_labs
```
